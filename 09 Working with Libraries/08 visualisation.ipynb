## Plotting a set of points

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['jan', 'Feb', 'Mar', 'Apr', 'May', 'June'])
y = np.array([500, 400, 800, 600, 300, 800])

plt.title("Expenditure per Month")
plt.xlabel("Month")
plt.ylabel("Amount OMR")


plt.plot(x, y)
plt.show()
```

## Sub-plot

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax= plt.subplots(2,2,figsize=(15,15))

y = np.array([500, 400, 800, 600, 300, 800])
x = np.array(['jan', 'Feb', 'Mar', 'Apr', 'May', 'June'])
lbl = ['jan', 'Feb', 'Mar', 'Apr', 'May', 'June']

ax[0][0].title.set_text('Pie')
ax[0][0].pie(y, labels = lbl )

ax[0][1].title.set_text('H-Bar')
ax[0][1].barh(x, y,color = 'r')

ax[1][0].title.set_text('Bar')
ax[1][0].bar(x, y,color = 'b')

ax[1][1].title.set_text('Bar')
ax[1][1].plot(x, y,color = 'g')
```
