# What is a function
A function is a block of code can only run when it is called.
You can pass data (parameters), into a function.
A function can return data as a result.


# Function format


# Types of functions
## 1. Does not take arguments and does not return values
## 2.	Takes arguments and does not return values
## 3.	Does not take arguments and returns values
## 4.	Takes arguments and returns values


### Type 1 Functions
This type of functions does not take any arguments and does not return any values when it is called. 


### Example: write a function that prints a welcome message to the screen when it is called

```python
def SayHi() :
    print ('Hi, welcome to Python')

SayHi()
```

### Type 2 Functions: 
This function takes arguments and does not return any values when it is called



### Ex: write a function that receives the name of a student and prints a welcome message to the screen when it is called

```python
def SayHi(name) :
  print ('welcome, ', name)

SayHi('Ali')
```

### Type 3 Functions: 
This function does not take arguments and returns values when it is called.


### Ex. A function which reads an integer and return it when it is called

```python
def Read_Int ():
  y = input ("Enter a number: ")
  y = int (y)
  return y


```

### Type 4 Functions: 
This function takes arguments and returns values when it is called


### Ex. Write a function that find whether an integer is odd or even

```python
def Even(x):
  z = x%2
  if z == 0:
    return True
  else: 
    return False

c = Even (11)
c    
```

### Same function with less code

```python

def Even(x):
  return (x%2==0)
Even(3)
```

## 03 Write a function the takes 2 numbers and returns the max and the minimum of the numbers

```python
def MinMax2(x, y):
  mx = 0
  mn = 0
  if x>y:
    mx = x
    mn = y
  elif y>x:
    mx = y
    mn = x
  else:
    mx = mn = x
  return mx, mn

a, b = MinMax2(-2,-2)
print ("Max = ", a, "Min = ", b)
```

## 04 Write a function the takes 3 numbers and returns the max and the minimum of the numbers

```python
def MinMax3(x, y, z):
  mx = 0
  mn = 0
  if x>y and x>z:
    mx = x
  elif y>x and y >z:
    mx = y
  else:
    mx= z

  if x<y and x<z:
    mn = x
  elif y<x and y <z:
    mn = y
  else:
    mn= z
  return mx, mn

a, b = MinMax3(8,4, -4)
print ("Max = ", a, "Min = ", b)  
```

## 09 Write a Python function to find the factorial of a number

```python

def fact(n):
  m = 1
  for i in range (1, n+1):
    m *=i
  return m

fact(3)
```

## 10 Write a function that finds the sum of all elements of a list

```python
def ListSum(L):
  s = 0
  for n in L:
    s+=n
  return s

A = [3, 5, 2,1]
x = ListSum(A)
x
```

## 11 Write a function that finds the average of all elements of a list

```python
def ListSum(L):
  s = 0
  for n in L:
    s+=n
  av = s/len(L)
  return av

A = [3, 5, 2,1]
x = ListSum(A)
x
```

## 13 Write a function that returns the max number and the min number in a list


```python
def MaxMinL(L):
  mx =L[0]
  mn = L[0]
  for n in L:
    if n>mx: 
      mx = n
    if n<mn:
      mn = n
  return mx, mn
A = [3, 12, -27]
x, y= MaxMinL(A)
print ("Max = ", x, " and Min = ", y)
```

## 14 write a function that receives two lists and return a list that contains the addition of the corresponding element

```python

def AddList(x, y):
  z = []
  for i in range(0, len(x)):
    r = x[i]+y[i]
    z.append(r)
  return z

A = [1, 2, 3, 8]
B = [3, 4, 5, 2]
C = AddList(A, B)
C
```

## 16 write a function that checks whether a number is prime


```python
def prime (n):
  cnt = 0
  for x in range (1, n+1):
    if n%x == 0:
      cnt +=1
  if cnt == 2:  return True
  else: return False

prime(121) 
```

## 19 Write a function that takes the GPA of a students and returns the Grade (fail, sat, good, very good, or excellent)​


```python
def Grade(mark):
  gr = ""
  if mark<50 and mark>=35:
    gr = 'Failed'
  elif mark >= 50 and mark<60:
    gr = "Pass"
  elif mark >=60 and mark<70:
    gr = "Satisfactory"
  elif mark >=70 and mark<80:
    gr = "Good"
  elif mark >=80 and mark<90:
    gr = "very good"
  elif mark >=90 and mark<=100:
    gr = "Excellent"
  else:
    gr = "Invalid"
  return gr

Grade(120)
```

## 24 Write a function that receives two number and an operation (+, -, *, /). The function returns the result according to the operation sent

```python
def Res (x, y, op):
    if op == "+":
        return x+y
    elif op == "*":
        return x*y
    elif op == "/":
        return x/y
    elif op == "-":
        return x-y
    else: return "Invalid Operation"
Res(1, 2, "/")
```
